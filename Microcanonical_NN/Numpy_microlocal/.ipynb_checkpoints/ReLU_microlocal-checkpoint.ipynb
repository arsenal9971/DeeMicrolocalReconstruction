# <center> Microlocal analysis of ReLU </center>

In this notebook we are analysing the microlocal properties of the ReLU function when seen as a point operator. $\text{ReLU}:L^2(\mathbb{R}^2)\longrightarrow L^2(\mathbb{R}^2)$ given by:

$$
\text{ReLU}(f)(x) = \text{ReLU}(f(x)) = \max\{0,f(x)\}
$$

## 1D 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = np.array([np.sin(x) for x in np.arange(0, 10, 0.2)])
```

```python
plt.plot(f)
```

```python
def ReLU(x):
    return np.array([max(xi,0) for xi in x])
```

```python
plt.plot(ReLU(f))
```

### Heaviside relation

The ReLU function can be seen as the multiplication $\text{ReLU}(y) = H(y) y$, therefore $\text{ReLU}(f)(x)=H(f(x))f(x)$ where

$$
H(f(x)) = \begin{cases}
f(x) & \text{ if } f(x)>0 \\
0 & \text{ if } f(x) \leq 0
\end{cases}
$$

```python
plt.figure(figsize = (6,6))
plt.axis("off")
plt.imshow(np.heaviside(f,0), cmap="gray")
```

## Gradient of heaviside function

We know that the weak gradient of $H(f)$ is given by the relation
$$
\int_{\mathbb{R}^2}\nabla H(f)(x)\cdot \varphi (x)dx = -\oint_{\partial \Omega} \varphi(x)\cdot n_{\partial\Omega}(x)dS(x)
$$

where $n_{\partial\Omega}(x)$ is the outward normal to the boundary surface $\partial\Omega$, where $\Omega= \text{supp}_+(f)$ (the non-negative support).

```python
from scipy import ndimage
```

```python
# Get x-gradient
fx = ndimage.sobel(f,axis=0,mode='constant')
# Get y-gradient
fy = ndimage.sobel(f,axis=1,mode='constant')
```

```python
# Get square root of sum of squares
sobel=np.hypot(fx,fy)
```

```python
plt.figure(figsize = (6,6))
plt.axis("off")
plt.imshow(sobel, cmap="gray")
```

```python
plt.figure(figsize = (6,6))
plt.axis("off")
plt.imshow(fx, cmap="gray")
```

### Computing the wavefront set

```python
WFimage = np.zeros(f.shape)
for i in range(f.shape[0]):
    for j in range(f.shape[1]):
        if fx[i,j] == 0:
            WFimage[i,j] = 0
        else:
            WFimage[i,j] = 180*np.arctan(fy[i,j]/fx[i,j])/(2*np.pi)
```

```python
plt.figure(figsize = (6,6))
plt.axis("off")
plot_WF(WFimage)
```
